```python
class Cliente:
    def __init__(self, rut, nombre, marca, modelo):
        self.rut = rut
        self.nombre = nombre
        self.marca = marca
        self.modelo = modelo
        self.servicios = []
        self.estado = "trabajando"

    def agregar_servicio(self, servicio):
        self.servicios.append(servicio)

    def actualizar_estado(self, estado):
        self.estado = estado

    def tiempo_total_reparacion(self, lavar_auto=False, lavado_gamuza=False):
        tiempos = {
            "Revision 1000 km": 2.0,
            "Cambio Aceite": 1.0,
            "Revision Frenos": 0.5,
            "Revision Correas": 0.5,
            "Revision Luces": 0.2,
            "Revision Direccion": 0.5
        }

        tiempo_total = 0
        for servicio in self.servicios:
            if servicio in tiempos:
                tiempo_total += tiempos[servicio]
            else:
                print(f"Servicio desconocido: {servicio}")

        if lavar_auto:
            tiempo_total += 0.5
        if lavado_gamuza:
            tiempo_total += 0.5
        
        return tiempo_total

    def mostrar_estado(self, lavar_auto=False, lavado_gamuza=False):
        print(f"\nCliente: {self.nombre} - Estado: {self.estado}")
        print("Servicios solicitados:")
        for servicio in self.servicios:
            print(f" - {servicio}")
        
        tiempo_total = self.tiempo_total_reparacion(lavar_auto, lavado_gamuza)
        if tiempo_total > 3:
            print(f"El tiempo total de reparación es largo: {tiempo_total} horas")
        elif tiempo_total > 1:
            print(f"El tiempo total de reparación es moderado: {tiempo_total} horas")
        else:
            print(f"El tiempo total de reparación es corto: {tiempo_total} horas")


def agregar_clientes():
    clientes = []
    agregar_mas = True

    while agregar_mas:
        rut = input("Ingrese el RUT del cliente: ")
        nombre = input("Ingrese el nombre completo del cliente: ")
        marca = input("Ingrese la marca del automóvil: ")
        modelo = input("Ingrese el modelo del automóvil: ")

        cliente = Cliente(rut, nombre, marca, modelo)

        agregar_servicios = True
        while agregar_servicios:
            servicio = input("Ingrese un servicio solicitado (o 'salir' para terminar): ")
            if servicio.lower() == 'salir':
                agregar_servicios = False
            else:
                cliente.agregar_servicio(servicio)

        lavar_auto = input("¿Desea lavar el auto? (si/no): ").lower() == 'si'
        lavado_gamuza = input("¿Desea lavado de gamuza? (si/no): ").lower() == 'si'

        tiempo_total = cliente.tiempo_total_reparacion(lavar_auto, lavado_gamuza)
        cliente.actualizar_estado("terminado" if tiempo_total > 0 else "trabajando")

        clientes.append(cliente)

        agregar_mas = input("¿Desea agregar otro cliente? (si/no): ").lower() == 'si'

    for cliente in clientes:
        cliente.mostrar_estado(lavar_auto, lavado_gamuza)


# Ejecutar la función para agregar clientes
agregar_clientes()
```
